```python
import pandas as pd
import numpy as np

# tabela = pd.read_csv('Radiação 159 pt.1.csv', encoding='latin-1', sep=';')
tabela = pd.read_excel(r'Radiação 159 pt.2.xlsx')
display(tabela)
```

```python
#SE DADOS VIEREM DIRETO DO REPORT GERADO NO NANODROP:
tabela = tabela.drop(['1 (nm)','#','User name','1 (Abs)'],axis=1)
display(tabela)
tabela.info()
```

```python
#SE DADOS VIEREM DIRETO DO REPORT GERADO NO NANODROP:
tabela['Date and Time'] = tabela['Date and Time'].astype('str')
tabela['Sample ID'] = tabela['Sample ID'].astype('str')
tabela['abs'] = tabela['abs'].astype('float')
```

```python
#SE DADOS VIEREM DIRETO DO REPORT GERADO NO NANODROP:
# tabela['Dia do mês'] = tabela['Date and Time'].str.split('/').str[0] #não utilizar
tabela['Data'] = tabela['Date and Time'].str[:5]
tabela['Dose (J)'] = tabela['Sample ID'].str.split().str[1]
display(tabela)
```

```python
#SE DADOS VIEREM DO SPECTRUM, APÓS PASSAREM PELA "Seleção de comprimentos de onda específicos com espectro NanoDrop"
# tabela = tabela.drop(['Unnamed: 0'], axis=1)
# tabela['Data'] = tabela['Data'].astype('str')
# tabela['Dose (J)'] = tabela['Dose (J)'].astype('str')
# display(tabela)
```

```python
tabela_nova = pd.DataFrame()
tabela0 = tabela.groupby('Dose (J)').get_group('0')
tabela100 = tabela.groupby('Dose (J)').get_group('100')
tabela300 = tabela.groupby('Dose (J)').get_group('300')
tabela500 = tabela.groupby('Dose (J)').get_group('500')
tabela700 = tabela.groupby('Dose (J)').get_group('700')
tabela1000 = tabela.groupby('Dose (J)').get_group('1000')
lista_tabelas = [tabela0,tabela100,tabela300,tabela500,tabela700,tabela1000]
doses = [0,100,300,500,700,1000]
for i,tabela in enumerate(lista_tabelas):
    dose = doses[i]
    tabela_nova['{}'.format(dose)] = tabela.groupby('Data')['abs'].mean()
    tabela_nova['{} desvio padrão'.format(dose)] = tabela.groupby('Data')['abs'].std()

display(tabela_nova)
```

```python
# tabela_nova.to_csv(r'Radiação 230 pt.2 - médias e desvios.csv',sep=';')
tabela_nova.to_excel(r'Radiação 159 - médias e desvios 2.xlsx')
```
